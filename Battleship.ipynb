```python
from random import randint

#Create a the Board as a list
board = []

# To create 5x5 grid
for x in range(5):
    #Using append to add each row
  board.append(["O"] * 5)

# Function to print the board row by row
def print_board(board):
  for row in board:
    # Used join to replace "," with " "
    print(" ".join(row))

#Calling Board function
print_board(board)

# Function to generate random row to hide the battleship
def random_row(board):
  return randint(0, len(board) - 1)

# Function to generate random column to hide the battleship
def random_col(board):
  return randint(0, len(board[0]) - 1)

# Saving random generated row and column value in variable ship_row and ship_col
ship_row = random_row(board)
ship_col = random_col(board)

# For debugging
print(ship_row)
print(ship_col)

# Giving 5 turns to the user to guess the location of the battleship
for turn in range(5):
  print("Turn", turn+1)
# Input row and column value from the user
  guess_row = int(input("Guess Row: "))
  guess_col = int(input("Guess Col: "))
# Check#1 to see if the value matches
  if guess_row == ship_row and guess_col == ship_col:
    print("Congratulations! You sunk my battleship!")
  else:
    # Check#2 to see whether the value entered by user is out of the ocean
    if (guess_row < 0 or guess_row > 4) or (guess_col < 0 or guess_col > 4):
      print("Oops, that's not even in the ocean.")
    # Check#3 to see if the value is previously entered by the user
    elif(board[guess_row][guess_col] == "X"):
      print("You guessed that one already.")
    # Check#4 when user fails to find the battleship
    else:
      print("You missed my battleship!")
    # Assigning the cell as "X" everytime the user makes a guess
      board[guess_row][guess_col] = "X"
  print_board(board)
```
